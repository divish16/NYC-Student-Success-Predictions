```python
import pandas as pd
import numpy
```

```python
file_1 = "2006-2012-school-demographics-and-accountability-snapshot.csv.zip"
file_2 = "2016-2017_Graduation_Outcomes_School.csv"
```

```python
demographics_df = pd.read_csv(file_1)
graduation_df = pd.read_csv(file_2)
```

```python
demographics_df
```

```python
drop_df = demographics_df.drop(columns = ["prek","k", "grade1", "grade2","grade3","grade4","grade5","grade6","grade7","grade8"])
drop_df
```

```python
school_demographics_df = drop_df.dropna()
school_demographics_df
```

```python
school_demographics_df.count()
```

```python
school_demographics_df.columns
```

```python
school_demographics_df.rename(columns = {"Name":"School Name"}, inplace = True)
school_demographics_df
```

```python
graduation_df.columns
```

```python
graduation_df
```

```python
merge_df = pd.merge(school_demographics_df,graduation_df, on=["School Name" , "School Name"])
merge_df
```

```python
merge_df.count()
```

```python
merge_drop_df = merge_df.drop(columns=['SACC (IEP Diploma) #',"SACC (IEP Diploma) % of cohort",'TASC (GED) #',
       'TASC (GED) % of cohort'])
merge_drop_df
```

```python
merge_drop_df.count()
```

```python
clean_data_df = merge_drop_df.dropna()
clean_data_df
```

```python
clean_data_df.count()
```

```python
# export pandas dataframe to csv
clean_data_df.to_csv('sample_data.csv')
```
